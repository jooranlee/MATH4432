```r
R_Squared = data.frame("Outstate - " = character(), r_squared = double())
j = 1

for (i in 1:ncol(college)){
    col = colnames(college)[i]
    if (col == "Outstate"){
        next
    }
    lm.fit = lm(college$Outstate ~ college[[i]])
    r_squared = summary(lm.fit) $ r.squared

    R_Squared[j,] = c(col, r_squared)
    j = j+1
}

R_Squared = R_Squared[ order(R_Squared[,2] ), ]
print(R_Squared)
```

One interesting discovery is that the number of applications, acceptance, and enrollment are among the variables with the least value of R^2.

```r
lm_Auto <- lm(mpg ~ .-name, data = Auto)
summary(lm_Auto)
```

By performing F-test with null hypothesis that all coefficient are zero, i.e. H0: beta_i = 0 for all i, the p-value is given by < 2.2e-16. This indicates that there exists relationship between the predictors and the response and we reject the null hypothesis.

From the p-value of the coefficient of the predictors, we can conclude that all predictors appear to have a statistically significant relationship to the response except cylinders, horsepower, and acceleration.

Some predictors, namely displacement, acceleration, year, and origin, have positive correlation with crim while others have negative correlation to crim, such as cylinders, horsepower, and weight. In other words, automobiles with higher displacement, acceleration, year, and origin tends to have higher mpg and automobiles with higher cylinders, horsepower, and weight tends to have lower mpg.

The coefficient for the year variable suggest that an increasing year variable by 1 while holding the other predictors fixed is associated with an average increase of 0.750773 in the mpg. This means that efficiency of automobiles increase by 0.750773/year.

```r
library(FNN)
library(ggplot2)
rm(list=ls())
set.seed(10)

n <- 200
nt <- n
nrep <- 100
p <- 20
puse <- c(1,2,3,4,10,20)
k <- c(1:9)

sigma <- 0.7

X <- replicate(p,runif(n,-2,2))
Xt <- X
y0 <- sin(2*X[,1])
yt <- sin(2*Xt[,1])

out_knn <- data.frame()
out_lm <- data.frame()

for(i in 1:length(puse)){
  yhat_lm <- matrix(0,nt,nrep)
  yhat_knn <- replicate(length(k),matrix(0,nt,nrep))
  for(l in 1:nrep){
    y <- y0 + rnorm(n,0,sigma)
    
    ######### DO: fit linear regression using lm funciton, assign predicted value to yhat_lm[,l] #########

    lm.fit = lm(y ~ X[, 1:i])
    yhat_lm[,l] <- predict(lm.fit, data.frame(Xt[, 1:i])) # predicted value by lm
      
    
    for(j in 1:length(k)){
      ######### DO: fit knn using knn.reg funciton, assign predicted value to yhat_knn[,l,j] #########

      yhat_knn[,l,j] <- knn.reg(as.data.frame(X[, 1:puse[i]]), as.data.frame(Xt[, 1:puse[i]]), y, k=k[j])$pred # predicted value by knn.reg
        
    }
    
    cat(i,"-th p, ",l,"-th repitition finished. \n")
  }
  
  ######### DO: compute bias and variance of linear regression #########
  # compute mean of predicted values
    
  ybar_lm <- rowMeans(yhat_lm) # E(f^hat)
  # compute bias^2
    
  biasSQ_lm <- mean((y0 - ybar_lm) ^ 2) # E[ (f - E(f^hat))^2 ]
    
  # compute variance
    
  variance_lm <- mean((yhat_lm - ybar_lm) ^ 2) # E[ (E(f^hat) - f^hat)^2 ]
    
  # compute total MSE
    
  err_lm <- biasSQ_lm + variance_lm
  
  out_lm <- rbind(out_lm,data.frame(error=biasSQ_lm,component="squared-bias",p=paste0("p=",puse[i])))
  out_lm <- rbind(out_lm,data.frame(error=variance_lm,component="variance",p=paste0("p=",puse[i])))
  out_lm <- rbind(out_lm,data.frame(error=err_lm,component="MSE",p=paste0("p=",puse[i])))
  
  
  ######### IMPLEMENT: compute bias and variance of knn regression #########
  # compute mean of predicted values
    
  ybar_knn <- colMeans(aperm(yhat_knn, c(2,1,3))) # E(f^hat)
    
  # compute bias^2
    
  biasSQ_knn <- colMeans((y0 - ybar_knn) ^ 2) # E[ (f - E(f^hat))^2 ]
    
  # compute variance
    
  variance_knn <- rowMeans(colMeans((aperm(yhat_knn, c(1,3,2)) - replicate(n = nrep, ybar_knn)) ^ 2)) # E[ (E(f^hat) - f^hat)^2 ]
    
  # compute total MSE
    
  err_knn <- biasSQ_knn + variance_knn
  
  out_knn <- rbind(out_knn,data.frame(error=biasSQ_knn,component="squared-bias",K=1/k,p=paste0("p=",puse[i])))
  out_knn<- rbind(out_knn,data.frame(error=variance_knn,component="variance",K=1/k,p=paste0("p=",puse[i])))
  out_knn<- rbind(out_knn,data.frame(error=err_knn,component="MSE",K=1/k,p=paste0("p=",puse[i])))
}
```

```r
options(repr.plot.width=10, repr.plot.height=3)
p <- ggplot(out_knn,aes(x=K,y=error,col=component)) + geom_line() + scale_x_log10() +
  geom_hline(data=out_lm,aes(yintercept=error,col=component),linetype="dashed") + 
  facet_grid(.~p) +
  labs(x="1/K",y="Error")
p
```
